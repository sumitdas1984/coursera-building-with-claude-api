```python
# Load env variables
from dotenv import load_dotenv
import os

load_dotenv
```

```python
# Create an API client instance
from anthropic import Anthropic

client = Anthropic()
```

```python
def add_user_message(messages, text):
    messages.append({"role": "user", "content": text})


def add_assistant_message(messages, text):
    messages.append({"role": "assistant", "content": text})     
    

def chat(messages):
    message = client.messages.create(
        model="claude-3-5-haiku-20241022",
        max_tokens=1000,
        messages=messages
    )
    return message.content[0].text
```

```python
# Make an initial list of messages
messages = []

# Use a 'while True' loop to keep the conversation going
while True:
    # Get user input
    user_input = input("> ")
    print(">:", user_input)
    
    # Add the user input to the list of messages
    messages.append({"role": "user", "content": user_input})
    
    # Call Claude with the 'chat' function
    answer = chat(messages)
    
    # Add the assistant response to the list of messages
    messages.append({"role": "assistant", "content": answer})
    
    # Print the assistant response
    print("Assistant:", answer)
```
